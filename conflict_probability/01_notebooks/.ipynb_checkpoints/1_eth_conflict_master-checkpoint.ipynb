```python
df = pd.read_csv('/Users/yaredhurisa/Google Drive/Probability of conflict/data/Conflict-Ethiopia.csv')
```

```python
df.isna().sum()
```

```python
df['month'] = pd.DatetimeIndex(df['event_date']).month
```

```python
event_type= df.groupby(['event_type', 'actor1', 'inter1']).count()
```

##### Actor code under inter1, inter2 and interaction
###### 1 = Governments and State Security Services
###### 2 = Rebel Groups
###### 3 = Political Militias
###### 4 = Identity Militias
###### 5 = Rioters
###### 6 = Protesters
###### 7 = Civilians
###### 8 = External/other forces


```python
#under interaction, select data representing the presence of identity militias (4)
cc=df[df['interaction'].isin([14, 24, 34, 41, 42, 43, 44, 45, 46, 47, 48])]
```

```python
actor1=pd.get_dummies(cc['inter1'])
actor1.head()
```

```python
actor2=pd.get_dummies(cc['inter2'])
actor2.head()
```

```python
actor=actor1.merge(actor2, left_index=True, right_index=True)
actor.head()
```

```python
actor['actor_state']=actor['1_x'] + actor['1_y']
actor['actor_political_militias']=actor['3_x'] + actor['3_y']
actor['actor_identity_militias']=actor['4_x'] + actor['4_y']
actor['actor_others']=actor['8_x'] + actor['8_y']
actor=actor.rename(columns={2: "actor_rebel_groups", 7: "actor_civilians"})
actor.head()
```

```python
actor['actor_identity_militias']=actor['actor_identity_militias'].replace([2], 1)
```

```python
actor=actor[['actor_state', "actor_rebel_groups", 'actor_political_militias', 'actor_identity_militias', "actor_civilians", 'actor_others']]
```

```python
actor.head()
```

```python
cc1=cc.merge(actor, left_index=True, right_index=True)
cc1.head()
```

```python
cc1['cc_event_count']=1
```

```python
cc2 = cc1.groupby(['year', 'month','admin1','admin2'])['fatalities', 'cc_event_count', 'actor_state', "actor_rebel_groups", 'actor_political_militias', 'actor_identity_militias', "actor_civilians", 'actor_others'].sum().reset_index()
```

```python
cc2 = cc2.set_index(['admin1','admin2','month', 'year'])
cc2['cc_onset']=1
```

```python
cc3=cc2.unstack().fillna(0).stack()
```

```python
cc3=pd.DataFrame(data=cc3)
```

```python
cc3=cc3.reset_index()
```

```python
cc3 = cc3.set_index(['admin1','admin2','year', 'month'])
cc3
```

```python
a = np.array(cc3['actor_state'].values.tolist())

cc3['actor_state'] = np.where(a > 0, 1, a).tolist()
```

```python
b = np.array(cc3['actor_political_militias'].values.tolist())

cc3['actor_political_militias'] = np.where(b > 0, 1, b).tolist()
```

```python
c = np.array(cc3['actor_identity_militias'].values.tolist())

cc3['actor_identity_militias'] = np.where(c > 0, 1, c).tolist()
```

```python
d = np.array(cc3['actor_others'].values.tolist())

cc3['actor_others'] = np.where(d > 0, 1, a).tolist()
```

```python
e = np.array(cc3['actor_rebel_groups'].values.tolist())

cc3['actor_rebel_groups'] = np.where(e > 0, 1, a).tolist()
```

```python
f = np.array(cc3['actor_civilians'].values.tolist())

cc3['actor_civilians'] = np.where(e > 0, 1, a).tolist()
```

```python
cc3=cc3.unstack().fillna(0).stack().reset_index()
```

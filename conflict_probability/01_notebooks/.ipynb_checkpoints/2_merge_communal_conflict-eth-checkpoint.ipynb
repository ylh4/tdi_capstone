### 5. Population

```python
pop=pop.replace({'admin2':{"Ilu Aba Bora":"Illubabor",
                         "Bahir Dar":"Bar Dar Sp. Zone",
                         "Zone 03":"Zone 3",
                         "Zone 02":"Zone 2",
                         "East Hararge":"East Harerghe",
                         "West Hararge":"West Harerghe",
                         "South Tigray":"Southern Tigray",
                         "Nuwer":"Nuer",
                         "Kefa":"Keffa",
                         "West Arsi":"Arsi",
                         "Jimma Sp.":"Jimma",
                         "Siliti":"Silti",
                         "Assosa":"Asosa",
                         "Dawuro":"Dawro",
                         "Zone 01":"Zone 1",
                         "Agnuwak":"Agnuak",
                         "Zone 04":"Zone 4",
                         "Kembata Timbaro":"Kembata Tembaro",
                         "West Gojjam":"West Gojam",
                         "North Western Tigray":"Western Tigray",
                         "South Wolo":"South Wollo",
                         "East Gojjam":"East Gojam",
                         "Horo Gudru Wellega":"Horo Gudru Welega",
                         "Zone 03":"Zone 3",
                         "Zone 02":"Zone 2",
                           "Wag Himra":"Wag Hemira",
                           "Bahir Dar":"Bar Dar Sp. Zone",
                           "Zone 03":"Zone 3",
                           "Zone 02":"Zone 2",
                           "Welwel & Warder":"Warder"
}})
```

```python
pop=pop.replace({'admin1':{"Gambella":"Gambela Peoples",
                         "Harari":"Harari People",
                         "Benishangul Gumuz":"Benshangul-Gumaz",
                         "Oromiya":"Oromia"
                         
}})
```

```python
from fuzzywuzzy import process
clist=df[['admin1']]
cdata=pop[['admin1']]
clist = clist.sort_values('admin1').drop_duplicates()
cdata = cdata.sort_values('admin1').drop_duplicates()

# create a choice list
choices = clist['admin1'].values.tolist()

# apply fuzzywuzzy to each row using lambda expression
cdata['Close Country'] = cdata['admin1'].apply(lambda x: process.extractOne(x, choices))

# add percent match wiht apply
cdata[['Close Country', 'Percent Match']] = cdata['Close Country'].apply(pd.Series)

cdata = cdata.sort_values('Percent Match')
cdata
```

```python
from fuzzywuzzy import process
clist=df[['admin2']]
cdata=pop[['admin2']]
clist = clist.sort_values('admin2').drop_duplicates()
cdata = cdata.sort_values('admin2').drop_duplicates()

# create a choice list
choices = clist['admin2'].values.tolist()

# apply fuzzywuzzy to each row using lambda expression
cdata['Close Country'] = cdata['admin2'].apply(lambda x: process.extractOne(x, choices))

# add percent match wiht apply
cdata[['Close Country', 'Percent Match']] = cdata['Close Country'].apply(pd.Series)

cdata = cdata.sort_values('Percent Match')
cdata = cdata[cdata['Percent Match']>70]
cdata
```

```python
pop=pop.groupby(['admin1', 'admin2', 'year',])['btotl'].sum().reset_index()
pop.head()
```

```python
df4=pd.merge(df3, pop, how='left', on=['admin1', 'admin2', 'year'])
df4.shape
```

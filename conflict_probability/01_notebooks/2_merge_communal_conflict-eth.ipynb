# Missing data handling using MICE


```python
impute_list=[
'inflation_all',
 'inflation_food',
 'inflation_non_food',
 'rainfall',
 'temperature mean',
 'btotl',
 'maize_ETB_KG',
 'teff_ETB_KG',
 'wheat_ETB_KG',
 'mean ndvi',
 'ndvi_lag',]
```

```python
# impute using MICE method
from impyute.imputation import cs
mice_impute=cs.mice(df7[impute_list].values)
df_mice=pd.DataFrame(mice_impute, columns=df7[impute_list].columns, 
                     index=df7.index)
```

```python
print(df_mice.isna().sum())
```

```python
df7[impute_list]=df_mice
print(df7.isna().sum())
```

```python
df7=df7.fillna(0)
```
